```python
# Separate features by type
numerical_features = train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = train.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
```

```python
from sklearn.impute import SimpleImputer

# Numerical - median imputation
num_imputer = SimpleImputer(strategy='median')
train[numerical_features] = num_imputer.fit_transform(train[numerical_features])

# Categorical - mode imputation
cat_imputer = SimpleImputer(strategy='most_frequent')
train[categorical_features] = cat_imputer.fit_transform(train[categorical_features])
```

```python
# List of categorical features
categorical_features = ['product_type', 'sub_area', 'culture_objects_top_25',
                        'thermal_power_plant_raion', 'incineration_raion', 'oil_chemistry_raion',
                        'radiation_raion', 'railroad_terminal_raion', 'big_market_raion',
                        'nuclear_reactor_raion', 'detention_facility_raion', 'water_1line',
                        'big_road1_1line', 'railroad_1line', 'ecology']

# Explore feature nature
feature_info = []

for col in categorical_features:
    n_unique = train[col].nunique()
    top_values = train[col].value_counts().head(5).to_dict()
    
    # Identify if likely binary
    if n_unique == 2:
        nature = 'Binary'
    elif n_unique <= 10:
        nature = 'Low-cardinality Categorical'
    else:
        nature = 'High-cardinality Categorical'
    
    feature_info.append({
        'Feature': col,
        'Unique_Values': n_unique,
        'Top_Values': top_values,
        'Nature': nature
    })

# Convert to DataFrame for nice display
feature_info_df = pd.DataFrame(feature_info)
feature_info_df
```

```python
binary_features = ['product_type', 'culture_objects_top_25', 'thermal_power_plant_raion',
                   'incineration_raion', 'oil_chemistry_raion', 'radiation_raion',
                   'railroad_terminal_raion', 'big_market_raion', 'nuclear_reactor_raion',
                   'detention_facility_raion', 'water_1line', 'big_road1_1line', 'railroad_1line']

for col in binary_features:
    train[col] = train[col].map({'no': 0, 'yes': 1, 'Type_A': 0, 'Type_B': 1, 'no':0,'yes':1})  # adjust for product_type if needed
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_features = ['ecology']
ecology_order = ['Eco_0', 'Eco_1', 'Eco_2', 'Eco_3', 'Eco_4']  # adjust according to real order
ord_enc = OrdinalEncoder(categories=[ecology_order])
train[ordinal_features] = ord_enc.fit_transform(train[ordinal_features])
```

```python
freq = train['sub_area'].value_counts() / len(train)
train['sub_area_encoded'] = train['sub_area'].map(freq)

target_mean = train.groupby('sub_area')['price_doc'].mean()
train['sub_area_encoded'] = train['sub_area'].map(target_mean)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Drop non-numeric or redundant columns
drop_cols = ['id', 'sub_area', 'ecology', 'price_doc']  # keep encoded versions
X = train.drop(columns=drop_cols)
y = train['price_doc']

# Align test data with same preprocessing
test_X = test.drop(columns=['id', 'sub_area', 'ecology'])

# Encode sub_area in test using same mapping
test_X['sub_area_encoded'] = test['sub_area'].map(target_mean)

# Fill any NaNs in test (e.g., unseen sub_areas) with global mean
test_X['sub_area_encoded'].fillna(y.mean(), inplace=True)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=29325
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error
import warnings
from sklearn.metrics import r2_score
warnings.filterwarnings('ignore')

# Preprocessing
num_features = X.select_dtypes(include=['int64', 'float64']).columns
cat_features = X.select_dtypes(include=['object', 'category']).columns

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
])

# Pipeline
lr_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

# Train & predict
lr_pipeline.fit(X_train, y_train)
y_pred = lr_pipeline.predict(X_val)

# RMSE
rmse_lr = mean_squared_error(y_val, y_pred, squared=False)
r2 = r2_score(y_val, y_pred)
print("Linear Regression RMSE:", rmse_lr)
print("Linear Regression RMSE on validation:", r2)
```

```python
# ============================================
# PREPROCESS TEST DATA (same as train)
# ============================================

# 1. Get numerical and categorical features (excluding target 'price_doc')
numerical_features_test = [col for col in numerical_features if col in test.columns]
categorical_features_test = [col for col in categorical_features if col in test.columns]

# 2. Impute missing values in test
num_imputer = SimpleImputer(strategy='median')
test[numerical_features_test] = num_imputer.fit_transform(test[numerical_features_test])

cat_imputer = SimpleImputer(strategy='most_frequent')
test[categorical_features_test] = cat_imputer.fit_transform(test[categorical_features_test])

# 3. Encode binary features in test
binary_features = ['product_type', 'culture_objects_top_25', 'thermal_power_plant_raion',
                   'incineration_raion', 'oil_chemistry_raion', 'radiation_raion',
                   'railroad_terminal_raion', 'big_market_raion', 'nuclear_reactor_raion',
                   'detention_facility_raion', 'water_1line', 'big_road1_1line', 'railroad_1line']

for col in binary_features:
    test[col] = test[col].map({'no': 0, 'yes': 1, 'Type_A': 0, 'Type_B': 1})

# 4. Encode ordinal feature in test
ecology_order = ['Eco_0', 'Eco_1', 'Eco_2', 'Eco_3', 'Eco_4']
ord_enc = OrdinalEncoder(categories=[ecology_order])
test['ecology'] = ord_enc.fit_transform(test[['ecology']])

# 5. Target encode sub_area in test (using train's target mean)
test['sub_area_encoded'] = test['sub_area'].map(target_mean)
test['sub_area_encoded'].fillna(y.mean(), inplace=True)

# 6. Prepare test features (same columns as X_train)
test_X = test[X_train.columns]
```

```python
# ============================================
# PREDICT ON TEST
# ============================================
y_test_pred = lr_pipeline.predict(test_X)

# ============================================
# CREATE SUBMISSION FILE
# ============================================
submission = pd.DataFrame({
    'id': test['id'].astype(int),
    'price_doc': y_test_pred
})
submission.to_csv('submission.csv', index=False)
print("✅ Submission file created: submission.csv")
print(f"Predictions shape: {submission.shape}")
print(submission.head())
```
